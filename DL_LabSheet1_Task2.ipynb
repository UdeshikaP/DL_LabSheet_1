```python
df = pd.read_csv('Au_nanoparticle_dataset (1).csv')
```

```python

# Create new dataframe with selected columns
selected_columns = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
new_df = df.drop(columns=selected_columns)
except_df = df[selected_columns]
```

```python

# Display first 20 samples
print("First 20 samples:\n")
new_df.head(20)
```

```python
#Calculate Mean
print("\nMean:")
except_df.mean()
```

```python
#Calculate Standard Deviation
print("\nStandard Deviation:")
except_df.std()
```

```python
#Calculate quartile
print("\nQuartiles:")
except_df.quantile([0.25, 0.5, 0.75])
```

```python
#Plot histograms in 1x4 layout
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 4))
for i, col in enumerate(selected_columns):
    plt.subplot(1, 4, i+1)
    sns.histplot(except_df[col], kde=True, color='red')
    plt.title(f'Histogram of {col}')
plt.tight_layout()
plt.show()
```

```python
#Pairplot visualization
sns.pairplot(except_df)
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

```python
#Custom PairGrid with KDE and histograms
g = sns.PairGrid(except_df)

# Bivariate histogram in upper triangle
g.map_upper(sns.histplot)

# Histogram + KDE in diagonal
g.map_diag(sns.histplot, kde=True)

# Bivariate KDE in lower triangle
g.map_lower(sns.kdeplot, fill=True)

plt.suptitle("Customized PairGrid", y=1.02)
plt.show()
```
