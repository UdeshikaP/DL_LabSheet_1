**2. Create a random 100000x1 array from the same exponential distribution and visualize the
distribution histogram along with uniform and normal distributions.**


```python
import numpy as np
import matplotlib.pyplot as plt #used for plotting graph

exp_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)

plt.hist(uniform_data, density=True, bins=200, histtype='step', color='blue', label='Uniform')
plt.hist(normal_data, density=True, bins=200, histtype='step', color='red', label='Normal')
plt.hist(exp_data, density=True, bins=200, histtype='step', color='green', label='Exponential')

plt.axis([0, 10, 0, 1.1])  # Adjust axis
plt.legend(loc='upper right')
plt.title(" Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

**3. Using matplotlib, plot the Z = X****2 + Y****2 in a 3D plot.Limit the x and y values to [-5, 5]. **




```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z = X^2 + Y^2')
plt.title("3D plot of Z = X^2 + Y^2")
plt.show()
```

**4. Calculate the pearson standard coefficient and spearman rank correlation for the [HP,
Attack, Defense, Sp. Ark, Sp. Def, Speed] features**


```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt #plotting

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Select relevant stats
stats = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

# Filter the DataFrame to only include those columns
stats_df = df[[col for col in df.columns if col.lower() in [s.lower() for s in stats]]]

# Pearson Correlation
pearson_corr = stats_df.corr(method='pearson')

# Spearman Correlation
spearman_corr = stats_df.corr(method='spearman')

# Plotting heatmaps with values
plt.figure(figsize=(12, 5))

# Pearson Heatmap
plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr)
plt.title('Pearson Correlation Heatmap')

# Spearman Heatmap
plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr)
plt.title('Spearman Correlation Heatmap')

plt.show()
```
